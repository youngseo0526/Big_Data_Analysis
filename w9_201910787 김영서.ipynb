### 문제 1) 서울특별시 개인서비스요금의 품목 별 최대, 최소가

#### 최대가

```python
df.groupBy(df.품목).max("가격(원)").show(100)
```

#### 최소가

```python
df.groupBy(df.품목).min("가격(원)").show(100)
```
